```python
%%writefile multi_model_app.py
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, recall_score,accuracy_score

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import gradio as gr

from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split
from surprise import accuracy

import streamlit as st
import pickle

# Load the models
with open('kmeans_model.pkl', 'rb') as f:
    kmeans_model = pickle.load(f)

with open('model_churn.pkl', 'rb') as f:
    model_churn = pickle.load(f)

with open('model.pkl', 'rb') as f:
    model_recco = pickle.load(f)

# Load product data for recommendations
df_products = pd.read_csv('products.csv')  # Ensure this path is correct

# Define a function for customer segmentation
def predict_customer_segmentation(input_data):
    # Make prediction logic here
    prediction = kmeans_model.predict(input_data)
    return prediction

# Define a function for reseller churn
def predict_reseller_churn(reseller_data):
    # Make prediction logic here
    prediction = model_churn.predict(reseller_data)
    return prediction

# Define a function for customer top recommendation
def get_top_recommendations(customer_id, n_recommendations=5):
    # Convert customer_id to integer
    try:
        customer_id = int(customer_id)
    except ValueError:
        return "Invalid Customer ID. Please enter a valid integer."

    # Check if customer exists in the data
    if customer_id not in df_products['CustomerID'].values:
        return "Customer ID not found."

    # Get top N product recommendations for the customer using the model
    recommendations = []
    for product_id in df_products['ProductID'].unique():
        # Predict the rating for each product
        prediction = model_recco.predict(customer_id, product_id)
        recommendations.append((product_id, prediction.est))

    # Sort recommendations by predicted rating and get the top N
    recommendations.sort(key=lambda x: x[1], reverse=True)
    top_recommendations = recommendations[:n_recommendations]

    # Format the output
    output = f"Top {n_recommendations} recommendations for Customer {customer_id}:\n"
    for i, (product_id, score) in enumerate(top_recommendations, 1):
        output += f"{i}. Product {product_id} with predicted score {score:.2f}\n"

    return output

# Streamlit app layout
st.title("Multi-Model Prediction Interface")

model_choice = st.selectbox("Choose a Model",
                             ("Customer Segmentation",
                              "Reseller Churn",
                              "Customer Top Recommendation"))

if model_choice == "Customer Segmentation":
    st.subheader("Customer Segmentation Model")
    # Input boxes for Recency, Frequency, and Monetary
    recency = st.number_input("Enter Recency", min_value=0, step=1)
    frequency = st.number_input("Enter Frequency", min_value=0, step=1)
    monetary = st.number_input("Enter Monetary Value", min_value=0.0, step=0.01)

    if st.button("Predict"):
        input_data = pd.DataFrame([[recency, frequency, monetary]], columns=["Recency", "Frequency", "Monetary"])
        prediction = predict_customer_segmentation(input_data)
        x=""
        if prediction[0] == 0:
            x="Occasional Buyers"
        elif prediction[0] == 1:
            x="Loyal Customers"
        else:
            x="May Churn"
        st.write(f"This Customer is in the {x} Segment")

elif model_choice == "Reseller Churn":
    st.subheader("Reseller Churn Model")
    # Input boxes for Recency, Frequency, and Monetary
    recency = st.number_input("Enter Recency", min_value=0, step=1)
    frequency = st.number_input("Enter Frequency", min_value=0, step=1)
    monetary = st.number_input("Enter Monetary Value", min_value=0.0, step=0.01)

    if st.button("Predict"):
        reseller_data = pd.DataFrame([[recency, frequency, monetary]], columns=["Recency", "Frequency", "Monetary"])
        prediction = predict_reseller_churn(reseller_data)
        st.write(f"This Reseller is {'likely to churn' if prediction[0] == 1 else 'not likely to churn'}.")

elif model_choice == "Customer Top Recommendation":
    st.subheader("Customer Top Recommendation Model")
    customer_id = st.text_input("Enter Customer ID")
    n_recommendations = st.slider("Number of Recommendations", min_value=1, max_value=10, value=5)

    if st.button("Get Recommendations"):
        if customer_id:
            result = get_top_recommendations(customer_id, n_recommendations)
            st.text_area("Recommendations", value=result, height=200)
        else:
            st.warning("Please enter a Customer ID.")
```
